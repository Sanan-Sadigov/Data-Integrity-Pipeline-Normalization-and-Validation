TASK 3:

Apply validation rules to your dataset using Pandera library.

Some of the rules are as following:

Dimensional Columns' ID versions - Status, Product Line, Customer, City, State, Postal Code, Country, Deal Size should not be null, should be in dataframe;

Monetary and quantity values should not be null and should be float and required;

ID and other numeric columns should be int and required;

Date columns should be in Date format

```python
testschema = DataFrameSchema({
    "ORDERNUMBER": Column(Int, nullable=False, required=True),
    "QUANTITYORDERED": Column(Int, nullable=False, required=True),
    "PRICEEACH": Column(Float, nullable=False, required=True),
    "ORDERLINENUMBER": Column(Int, nullable=False, required=True),
    "SALES": Column(Float, nullable=False, required=True),
    "ORDERDATE": Column(DateTime, nullable=False, required=True),
    "QTR_ID": Column(Int, nullable=False, required=True),
    "MONTH_ID": Column(Int, nullable=False, required=True),
    "YEAR_ID": Column(Int, nullable=False, required=True),
    "MSRP": Column(Int, nullable=False, required=True),
    "PRODUCTCODE": Column(String, nullable=False, required=True),
    "PHONE": Column(String, nullable=False, required=True),
    "ADDRESSLINE1": Column(String, nullable=False, required=True),
    "ADDRESSLINE2": Column(String, nullable=True), 
    "TERRITORY": Column(String, nullable=True), 
    "CONTACTLASTNAME": Column(String, nullable=False, required=True),
    "CONTACTFIRSTNAME": Column(String, nullable=False, required=True),
    "Status_ID": Column(Int, nullable=False, required=True),
    "Product_ID": Column(Int, nullable=False, required=True),
    "Customer_ID": Column(Int, nullable=False, required=True),
    "City_ID": Column(Int, nullable=False, required=True),
    "State_ID": Column(Int, nullable=False, required=True),  
    "Postal_ID": Column(Int, nullable=False, required=True), 
    "Country_ID": Column(Int, nullable=False, required=True),
    "DealSize_ID": Column(Int, nullable=False, required=True)}, index=Index(Int), strict=True, coerce=True)

errdata = None
try:
    testschema.validate(merged_dataset, lazy=True)
except pa.errors.SchemaErrors as err:
    errfail = pd.DataFrame(err.failure_cases)
    errfail.to_csv('foundedd_errors.csv')

    errdata = pd.DataFrame(err.data)
    errdata.to_csv('something.csv',index=False)

    errorlist = errfail.merge(errdata, left_on='index', right_index=True)
    errorlist.to_csv('aaa_errorlist.csv')
```
